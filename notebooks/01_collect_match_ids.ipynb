```python
player_list = bronze4_players.puuid.tolist()
```

```python
import os

# Dosya adı:
API_KEY_FILE = "riot_secret.txt" 

try:

    with open(API_KEY_FILE, 'r') as f:
        API_KEY = f.read().strip()
        
        
except FileNotFoundError:
    
    API_KEY = None 
    
except Exception as e:
   
    API_KEY = None


HEADERS = {"X-Riot-Token": API_KEY}
```

```python
import requests
import time
import pandas as pd

headers = {"X-Riot-Token": API_KEY}
```

```python
def get_match_ids(puuid, count=5):
    url = f"https://americas.api.riotgames.com/lol/match/v5/matches/by-puuid/{puuid}/ids?queue=420&count={count}"
    
    r = requests.get(url, headers=headers)
    
    if r.status_code == 200:
        return r.json()       
    else:
        print("Hata:", r.status_code, r.text)
        return []            
```

```python
all_match_ids = []

for i, puuid in enumerate(player_list, start=1):

    
    if len(all_match_ids) >= 1500:
        print("✔ Limit reached: 1500 match IDs collected.")
        break

    ids = get_match_ids(puuid, count=5)
    all_match_ids.extend(ids)

    print(f"{i}. player → {len(ids)} match id collected total: {len(all_match_ids)}")

    time.sleep(1.3)
```

```python
len(all_match_ids)
```

```python
unique_match_ids = list(set(all_match_ids))
len(unique_match_ids)
```

```python
import json

with open("unique_match_ids.json", "w") as f:
    json.dump(unique_match_ids, f)
```

```python
import random

sample_size = 1000
sample_match_ids = random.sample(unique_match_ids, sample_size)
len(sample_match_ids)
```

```python
import requests
import time
import json

headers = {"X-Riot-Token": API_KEY}
```

```python
def get_match_detail(match_id):
    url = f"https://americas.api.riotgames.com/lol/match/v5/matches/{match_id}"
    r = requests.get(url, headers=headers)

    if r.status_code == 200:
        return r.json()          
    else:
        print("Hata:", match_id, r.status_code, r.text[:100])
        return None            
```

```python
all_match_info = []

for i, match_id in enumerate(sample_match_ids, start=1):

    detail = get_match_detail(match_id)

    if detail is not None:
        all_match_info.append(detail)   
        print(f"{i}. maç alındı → toplam: {len(all_match_info)}")
    else:
        print(f"{i}. maç alınamadı, atlandı.")

    time.sleep(1.2)   # RIOT rate-limit'e girmemek için bekleme

```

```python
import json

with open("bronze4_matches_all.json", "w", encoding="utf-8") as f:
    json.dump(all_match_info, f, ensure_ascii=False, indent=2)
```
